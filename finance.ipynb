```python
import matplotlib.ticker as ticker
import numpy as np
import pandas as pd
import seaborn as sns

from finance.values import compound_interest

%config InlineBackend.figure_formats = ['svg']
sns.set_theme(rc={"axes.formatter.limits": (-8, 9)})
```

```python
principal = 400000  # INR
roi_annual = 12 / 100  # percent
freq_per_year = 4
years = 20
```

```python
time = pd.Series(np.linspace(0, years, years * 2), name="time")
returns_over_time = pd.Series(
    compound_interest(principal, roi_annual, freq_per_year, time),
    name="returns",
)
```

```python
g = sns.relplot(x=time, y=returns_over_time, kind="line")


g.figure.suptitle(
    f"Compound Interest @ {roi_annual * 100} percent per annum, compounded {freq_per_year} times per year",
    fontsize=12,
)
g.figure.subplots_adjust(top=0.9)
g.ax.get_yaxis().set_major_formatter(
    ticker.FuncFormatter(lambda y, p: format(int(y), ","))
)
```
